```python
import matplotlib.pyplot as plt
import numpy as np
seasons = ['18-19', '19-20', '20-21']
G60 = [0.66, 0.75, 0.43]
ixG60 = [0.56, 0.49, 0.55]
plt.figure(figsize=(15, 7))
plt.subplot(131)

plt.plot(seasons, G60, color="red", label='G/60')
plt.scatter(seasons, G60, color="red")

plt.plot(seasons, ixG60,'--', color="#88d6e2", label='ixG/60')
plt.scatter(seasons, ixG60, color="#88d6e2")

plt.legend(title='Legend')

plt.grid(True)
plt.title('Jesperi Kotkaniemi\'s GAx/60\nin each of his first 3 seasons')
plt.suptitle('Data : Evolving-Hockey',x=0.3, y=-(1/1000000000000000000))
plt.savefig('KKGAx.png', dpi=500, facecolor = 'White', bbox_inches = "tight")
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator as ML
from matplotlib.ticker import ScalarFormatter as SF

fig, ax = plt.subplots()
X = np.linspace(0, 5, 100)
Y1, Y2 = np.sin(X), np.cos(X)



ax.xaxis.set_minor_locator(ML(0.2))
ax.xaxis.set_minor_formatter(SF())
ax.tick_params(axis='x', which='minor', rotation=90)

ax.plot(X, Y1, X, Y2)
```
